```python
"""
1. Set up a random experiment to test the difference between a sequential search and
a binary search on a list of integers
"""
"""
Process thinking:
main functions:
- a generator of sorted list of integer number between 0, 1000000
- Timer
- sequential_search
- binary_search
"""
import random, time

def time_it(func):
    def wrapper(*args, **kargs):
        start_time = time.perf_counter_ns()
        result = func(*args, **kargs)
        end_time = time.perf_counter_ns()
        elapsed_time = end_time - start_time
        print(f"==> Function '{func.__name__}' took {elapsed_time} nanoseconds to execute.")
        return result
    return wrapper

def generator_list(size_of_list: int) -> list:
    res: list[int]= []
    for num in range(size_of_list):
        res.append(random.randint(0,1000000))
    res.sort()
    return res
@time_it
def sequential_search(num_found: int, list_of_numbers: list[int])-> bool:
    for num in list_of_numbers:
        if num_found == num:
            return True
    return False

@time_it
def binary_search(num_found: int, list_of_numbers: list[int]) -> bool:
    length_of_list: int = len(list_of_numbers)
    new_len: list = list_of_numbers
    low: int = 0
    high: int = length_of_list
    while low < high:
        mid: int = (low + high)//2
        if list_of_numbers[mid] == num_found:
            return True
        elif list_of_numbers[mid] < num_found:
            low = mid+1
        else:
            high = mid
    return False

def simulator(num_of_simulations: int, num_of_elements: int) -> None:
    nums: list[int] = generator_list(num_of_elements)

    for i in range(num_of_simulations):
        num_to_search: int = random.randint(0, 1000000)
        print(f"num being searched: {num_to_search}")
        sequential_bool: bool = sequential_search(num_to_search, nums)
        binary_bool: bool = binary_search(num_to_search, nums)
        if (sequential_bool and binary_bool):
            print(f"num: {num_to_search} was found")
        else:
            print(f"num: {num_to_search} UNFOUND")

simulator(5, 10000)
```

```python
"""
2. Use the binary search functions given in the text(recursive and iterative). 
Generate a random, ordered list of integers and do a benchmark analysis for each one.
what are you result? Can you explain them?

Result: the binary search iterative performs better than a binary search recursive.
Explain: it is due to space time where the iterative tends to have a constant space time, 
the binary search recursive will tend to increase due to the creation of sub-function(stack) for each new slice to solve the sub-problems.
"""
import random, timeit

def generator_list(size_of_list: int) -> list:
    res: list[int]= []
    for num in range(size_of_list):
        res.append(random.randint(0,1000000))
    res.sort()
    return res

def binarySearchRecursive(alist, item):
    if len(alist) == 0:
        return False
    else:
        midpoint = len(alist)//2
        if alist[midpoint]==item:
          return True
        else:
          if item<alist[midpoint]:
            return binarySearchRecursive(alist[:midpoint],item)
          else:
             return binarySearchRecursive(alist[midpoint+1:], item)


def binarySearchIterative(alist, item):
    first = 0
    last = len(alist)-1
    found = False

    while first <= last and not found:
        midpoint = (first + last)//2
        if alist[midpoint] == item:
           found = True
        else:
            if item < alist[midpoint]:
               last = midpoint-1
            else:
               first = midpoint+1
    return found

def printSeconds(datas: list):
   ind: int = 0
   for data in datas:
    ind +=1
    print(f"{ind}: it took {data} seconds")
    
data_list: list = generator_list(10000)
numFound = random.randint(0,1000000)

TEST_CODE_RECURSIVE = '''
binarySearchRecursive(data_list, numFound)

''' 
TEST_CODE_ITERATIVE = ''' 
binarySearchIterative(data_list, numFound)
'''

times_recursive = timeit.repeat(stmt=TEST_CODE_RECURSIVE, globals=globals(), repeat=10, number=10000)
times_iterative = timeit.repeat(stmt=TEST_CODE_ITERATIVE, globals=globals(), repeat=10, number=10000)

print("# Recursive Function")
print()
printSeconds(times_recursive)
print()

print("# Iterative Function")
print()
printSeconds(times_iterative)
 
```
