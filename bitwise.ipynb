```python
# learning bitwise operator

##  we are going to use the following number
## x=7=111 and y=4=100
## x=10=1010 and y=4=0100

## Bitwise AND operator()
a = 7 
b = 4
c = -10

### Print bitwise AND operation
print(f"a<=>{a}<=>({bin(a)}) & b<=>{b}<=>({bin(b)}) = {a & b} <=> {bin(a & b)}")

## Bitwise OR Operator ()
print(f"a<=>{a}<=>({bin(a)}) | b<=>{b}<=>({bin(b)}) = {a | b} <=> {bin(a | b)}")

## Bitwise XOR Operator
print(f"a<=>{a}<=>({bin(a)}) ^ b<=>{b}<=>({bin(b)}) = {a ^ b} <=> {bin(a ^ b)}")

## Bitwise NOT Operator (Works with a single value and returns its on's complement)
print(f"~a<=>{a}<=>({bin(a)}) = {~a}<=>{bin(~a)}")
print(f"~b<=>{b}<=>({bin(b)}) = {~b}<=>{bin(~b)}")

# Bitwise Shift( shift the bits of a number left or right thereby multiplying or dividing the number by two respectively)

## Bitwise Right shift(shifts the bits of the number of the right and fills 0 on voids left or fills 1 in the case of a negative number)
## right shift has a similar effect as of dividing the number with some power of two
## operand >> number of shift to perform on the operand
print(f"a<=>{a}<=>({bin(a)}) >> 1 = {a >> 1} <=> {bin(a >> 1)}")
print(f"b<=>{b}<=>({bin(b)}) >> 1 = {b >> 1} <=> {bin(b >> 1)}")
print(f"c<=>{c}<=>({bin(c)}) >> 1 = {c >> 1} <=> {bin(c >> 1)}")
print(f"c<=>{c}<=>({bin(c)}) >> 2 = {c >> 2} <=> {bin(c >> 2)}")

## Bitwise left shift(Shifts the bits of number to the left and fills 0 on voids rights as result.)
## left shift has a similar effect as of multiplying the number with some power of two.
## operand << number of shift to perform on the operand

print(f"a<=>{a}<=>({bin(a)}) << 1 = {a << 1} <=> {bin(a << 1)}")
print(f"b<=>{b}<=>({bin(b)}) << 1 = {b << 1} <=> {bin(b << 1)}")
print(f"c<=>{c}<=>({bin(c)}) << 1 = {c << 1} <=> {bin(c << 1)}")
print(f"c<=>{c}<=>({bin(c)}) << 2 = {c << 2} <=> {bin(c << 2)}")

## Bitwise Operator Overloading
"""
Operator Overloading means giving extended meaning beyond their predefined operational meaning.
For instance + can add to int as well as concatenate two strings, it is because the + operator 
is overloaded by the int class and str class.

You have noticed that same built-in function or operator behave differently for object of
differents classes, this is called Operator Overloading
"""

class Bit_wise():
    def __init__(self, value):
        self.value = value

    def __and__(self, obj):
        print("AND & operator in bitwise operation")
        if isinstance(obj, Bit_wise):
            return self.value & obj.value
        else:
            raise ValueError("Must be an object of class Bit_wise")

    def __or__(self, obj):
        print("OR | operator in bitwise operation")
        if isinstance(obj, Bit_wise):
            return self.value | obj.value
        else:
            raise ValueError("Must be an object of class Bit_wise")

    def __xor__(self, obj):
        print("XOR ^ operator in bitwise operation")
        if isinstance(obj, Bit_wise):
            return self.value ^ obj.value
        else:
            raise ValueError("Must be an object of class Bit_wise")
    
    def __lshift__(self, obj):
        print("left Shift << operator in bitwise operation")
        if isinstance(obj, Bit_wise):
            return self.value << obj.value
        else:
            raise ValueError("Must be an object of class Bit_wise")

    def __rshift__(self, obj):
        print("left Shift >> operator in bitwise operation")
        if isinstance(obj, Bit_wise):
            return self.value >> obj.value
        else:
            raise ValueError("Must be an object of class Bit_wise")
    
    def __invert__(self):
        print("Invert operator represented the NOT ~ bit wise operator")
        return ~self.value
    
# main function 

if __name__ == "__main__":
    a = Bit_wise(10)
    b = Bit_wise(12)
    print(a & b)
    print(a | b)
    print(a ^ b)
    print(a << b)
    print(a >> b)
    print(~a)
```
