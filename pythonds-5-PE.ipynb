```python
"""
2. Write a recursive function to reverse a list.
"""
reversed_list = []
def rec_reverse(list_to_reverse):
    # base case: list of one elements, it is already reverse.
    if len(list_to_reverse) == 1:
         reversed_list.append(list_to_reverse.pop())
         return reversed_list
    else:
        reversed_list.append(list_to_reverse.pop())
        rec_reverse(list_to_reverse)
        return reversed_list

rec_reverse([1, 2, 3, 4 ,5 , 6, 7])
```

```python
"""
4. Find or invent an algorithm for drawing a fractal mountain. 
   Hint: One approach to this uses triangles again.
"""
import turtle

import turtle
import random

def mountain(P1, P2, roughness):
    # Base case: if segment is too small, draw a line
    # print(f"P1 = {P1}- P2 = {P2}")
    # print(f"{(P2[0] - P1[0])**2 + (P2[1] - P1[1])**2 }")
    if (P2[0] - P1[0])**2 + (P2[1] - P1[1])**2 < 2:
     #   print("passed")
        turtle.goto(P2)
        return

    # Calculate midpoint
    mid_x = (P1[0] + P2[0]) / 2
    mid_y = (P1[1] + P2[1]) / 2
    #print(f"{mid_x} - {mid_y}")

    # Apply random vertical displacement
    displacement = random.uniform(-roughness, roughness)
    mid_y += displacement

    P3 = (mid_x, mid_y)

    # Recursive calls for the two new segments
    mountain(P1, P3, roughness * 0.5)  # Reduce roughness for next level
    mountain(P3, P2, roughness * 0.5)

# Setup Turtle
turtle.speed(0) # Fastest drawing speed
turtle.penup()
turtle.goto(-500, 0) # Starting point of the mountain range
turtle.pendown()

# Initial call to draw the mountain
start_point = (-400, 0)
end_point = (400, 0)
initial_roughness = 100 # Adjust for desired level of detail

mountain(start_point, end_point, initial_roughness)

turtle.done()
turtle.exitonclick()
```

```python
def get_two_gallons(jug4, jug3, target, steps):
    # Print the current state of the jugs
    print(f"4-gallon jug: {jug4}, 3-gallon jug: {jug3}, steps: {steps}")
    
    # Check if we have reached the target
    if jug4 == target:
        return True
    
    # If not, we can try different operations
    # 1. Fill the 4-gallon jug
    if get_two_gallons(4, jug3, target, steps + 1):
        return True
    
    # 2. Fill the 3-gallon jug
    if get_two_gallons(jug4, 3, target, steps + 1):
        return True
    
    # 3. Empty the 4-gallon jug
    if get_two_gallons(0, jug3, target, steps + 1):
        return True
    
    # 4. Empty the 3-gallon jug
    if get_two_gallons(jug4, 0, target, steps + 1):
        return True
    
    # 5. Pour from 4-gallon jug to 3-gallon jug
    if jug4 > 0 and jug3 < 3:
        transfer = min(jug4, 3 - jug3)
        if get_two_gallons(jug4 - transfer, jug3 + transfer, target, steps + 1):
            return True
            
    # 6. Pour from 3-gallon jug to 4-gallon jug
    if jug3 > 0 and jug4 < 4:
        transfer = min(jug3, 4 - jug4)
        if get_two_gallons(jug4 + transfer, jug3 - transfer, target, steps + 1):
            return True
    
    return False

# Start the recursion with both jugs empty
if __name__ == "__main__":
    target_amount = 2
    print("Solving the jug problem:")
    if not get_two_gallons(0, 0, target_amount, 0):
        print("No solution found.")
```
