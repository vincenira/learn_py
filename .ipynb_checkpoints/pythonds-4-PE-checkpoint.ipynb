```python
# Chapter 4: programming exercises
"""
https://github.com/mbhushan/ps-algods-py/blob/master/chap3-BasicDataStructure/QueueLinkedListClient.py
1. Modify the infix-to-postfix algorithm so that it can handle errors.
Error:
- Wrong characters
- Wrong sequences of parentheses
Remaining, it is to implement errors.
"""
from pythonds.basic import Stack

def infixToPostfix(infixToken):
    prec = {}
    prec["*"] = 3 
    prec["/"] = 3
    prec["-"] = 2
    prec["+"] = 2
    prec["("] = 1
    opStack = Stack()
    postFix = []
    tokens = infixToken.split()
    syntaxChecker(tokens)
    for token in tokens:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "1234567890":
            postFix.append(token)
        elif token == "(":
            opStack.push(token)
        elif token == ")":
            topToken = opStack.pop()
            while topToken != "(":
                postFix.append(topToken)
                topToken = opStack.pop()
        else: 
            while (not opStack.isEmpty()) and (prec[opStack.peek()] >= prec[token]):
                postFix.append(opStack.pop())
            opStack.push(token)
    while not opStack.isEmpty():
        postFix.append(opStack.pop())
    return " ".join(postFix)
    
def syntaxChecker(tokens):
    parStack = Stack()
    for token in tokens:
        if token not in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" and token not in "1234567890" and token not in "*/-+()":
            raise SyntaxError(f'Could not recognize {token}!')
        if token == "(":
            parStack.push(token)
        if token == ")" and not parStack.isEmpty():
            parStack.pop()
    if not parStack.isEmpty():
        raise SyntaxError(f' Parenthese are not balanced!')
    
print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A * B + C * D"))
```

```python
"""
2. Modify the postfix evaluation algorithm so that it can handle errors.
Error
- Wrong characters
- Can't perform the calculation throw an error.
"""
from pythonds.basic import Stack

def postFixEval(postFixToken):
    tokens = postFixToken.split()
    operandStack = Stack()
    for token in tokens:
        if not token in "*-/+":
            operandStack.push(token)
        else:
            if not operandStack.isEmpty():
                try:
                    op2 = int(operandStack.pop())
                    op1 = int(operandStack.pop())
                except:
                    raise SyntaxError(f'Can not cast the operand!')
                result = doMath(op1,op2,token)
                operandStack.push(result)
    ret_result = operandStack.pop()
    if not operandStack.isEmpty():
        print("Did not work")
    return ret_result
        
def doMath(op1,op2,op3):
    if op3 == "*":
        result = op1 * op2
    elif op3 == "/":
        result = op1 / op2
    elif op3 == "+":
       result = op1 + op2
    else:
        result = op1 - op2
    return result
print(postFixEval('7 8 + 3 2 + /'))
```

```python
"""
3. Implement a direct infix evaluator that combines the functionality of infix-to-postfix conversion and the postfix evaluation algorithm. 
   Your evaluator should process infix tokens from left to right and use two stacks, one for operators and one for operands, to perform the evaluation.
   
   Note:First version is working but it will be a good exercise to see if i can dry the algorithms and not working for numbers greater than 9
"""
from pythonds.basic import Stack

def infix_evaluator(infix_token):
    prec = {}
    prec["*"] = 3 
    prec["/"] = 3
    prec["-"] = 2
    prec["+"] = 2
    prec["("] = 1
    opStack = Stack()
    opdStack = Stack()
    tokens = infix_token.split()
    for token in tokens:
        if token in "1234567890":
            opdStack.push(int(token))
        elif token == "(":
            opStack.push(token)
        elif token == ")":
            top_token = opStack.pop()
            while top_token != "(":
                op1 = opdStack.pop()
                op2 = opdStack.pop()
                result = doMath(op1, op2, top_token)
                opdStack.push(result)
                top_token = opStack.pop()        
        else: 
            while (not opStack.isEmpty()) and (prec[opStack.peek()] >= prec[token]):
                op1 = opdStack.pop()
                op2 = opdStack.pop()
                top_token = opStack.pop()
                result = doMath(op1, op2, top_token)
                opdStack.push(result)
            opStack.push(token)
    while not opStack.isEmpty():
        op1 = opdStack.pop()
        op2 = opdStack.pop()
        top_token = opStack.pop()
        result = doMath(op2, op1, top_token)
        opdStack.push(result)
    return opdStack.peek()      
            
def doMath(op1,op2,op3):
    if op3 == "*":
        result = op1 * op2
    elif op3 == "/":
        result = op1 / op2
    elif op3 == "+":
       result = op1 + op2
    else:
        result = op1 - op2
    return result

print(infix_evaluator("2 * ( 3 + 4 ) - 2"))
print(infix_evaluator("2 + 1"))
print(infix_evaluator("2 + 1 * 7 + 1"))
print(infix_evaluator("( ( 3 + 1 ) / 4 ) * 8 - 1 "))
print(infix_evaluator("2 + 1"))
```

```python
"""
You are absolutely right. It is indeed possible to achieve an average O(1) time complexity for enqueue and dequeue operations in a queue implementation,
with a caveat. 
This means that most of the time, these operations will be very fast (constant time), 
but there can be a specific scenario where dequeue becomes O(n).

Here's a common approach to achieve this:

**Resizing Array Implementation**

1. Use a regular Python list to store the queue elements.
2. Maintain two variables: `front` (index of the first element) and `rear` (index of the last element + 1).
3. **Enqueue (O(1) on average):** Append the new element to the end of the list and update `rear`. 
   If `rear` reaches the end of the list, resize the list to double its capacity (this is the occasional O(n) operation).
4. **Dequeue (O(1) on average):**  Remove the element at the `front` index and update `front`.  
    If the number of elements falls below a certain threshold (e.g., ¼ of the list capacity), 
    resize the list to half its capacity (another occasional O(n) operation).

**Why is dequeue O(n) in some cases?**

Resizing the list to accommodate new elements or shrink it when elements are dequeued can take O(n) time in the worst case, 
as all elements need to be copied to a new array. However, this resizing only happens occasionally, and most of the time, 
enqueue and dequeue will only modify the front or rear pointers (O(1)).

**Trade-offs**

This approach offers an average O(1) time complexity for enqueue and dequeue,
but it introduces some overhead due to potential resizing operations. 
It's a good choice when you prioritize constant time for most operations and occasional resizing is acceptable.

**Alternative approaches**

* **`collections.deque`:** As mentioned before, `collections.deque` offers guaranteed O(1) for enqueue and dequeue, 
but it might not be suitable if you need to track the size of the queue elements or have other specific requirements.
* **Circular Array (fixed size):** This approach provides constant time for enqueue and dequeue but has a limitation:
the queue size needs to be fixed beforehand.

The best approach depends on your specific needs and priorities. 
Consider factors like queue size, operation frequency, and whether occasional O(n) dequeues are acceptable.
"""
```

```python
import turtle
import time
def tree(branchLen,t):
    if branchLen > 5:
        t.forward(branchLen)
        t.right(20)
        time.sleep(5)
        tree(branchLen-15,t)
        t.left(40)
        time.sleep(5)
        tree(branchLen-15,t)
        t.right(20)
        time.sleep(5)
        t.backward(branchLen)

def main():
    t = turtle.Turtle()
    myWin = turtle.Screen()
    t.left(90)
    t.up()
    t.backward(100)
    t.down()
    t.color("green")
    tree(30,t)
    myWin.exitonclick()

main()
```
