```python
from pythonds.basic import Stack
"""Check if the parentheses closes properly"""
def check_parentheses(token):
    stack = Stack()
    balanced = True
    index = 0
    while index < len(token):
        if token[index] == "(":
            stack.push(token[index])
        else:
                
            if not stack.isEmpty():
                stack.pop()
            else:
                balanced = False
        index +=1
    return stack.isEmpty() and balanced
    
print(check_parentheses("(()()()())"))

print(check_parentheses("(((())))"))

print(check_parentheses("(()((())()))"))
print("==============================")
print(check_parentheses("((((((())"))

print(check_parentheses("()))"))

print(check_parentheses("(()()(()"))

    
```

```python
from pythonds.basic import Stack
"""Check if the parentheses closes properly"""
def parChecker(token):
    stack = Stack()
    balanced = True
    index = 0
    while index < len(token):
        if token[index] in "({[":
            stack.push(token[index])
        else:
            if not stack.isEmpty() and matches(stack.peek(), token[index]):
                stack.pop()
            else:
                balanced = False
        index +=1
    return stack.isEmpty() and balanced
def matches(open, close):
    opens="({["
    closes=")}]"
    return opens.index(open) == closes.index(close)

print(parChecker("{{([][])}()}"))
print(parChecker("[[{{(())}}]]"))
print(parChecker("[][][](){}"))
print(parChecker("([)]"))
print(parChecker("((()]))"))
print(parChecker("[{()]"))
```

```python
from pythonds.basic import Stack

def binaryConverter(decimal):
    binaryholder = Stack()
    while decimal > 0:
        rem = decimal % 2
        binaryholder.push(rem)
        decimal = decimal // 2
    binstring = ""
    while not binaryholder.isEmpty():
        binstring = binstring + str(binaryholder.pop())
    return binstring

def baseConverter(decimal, base):
    digits = "0123456789ABCDEF"
    binaryholder = Stack()
    while decimal > 0:
        rem = decimal % base
        binaryholder.push(rem)
        decimal = decimal // base
    binstring = ""
    while not binaryholder.isEmpty():
        binstring = binstring + digits[binaryholder.pop()]
    return binstring
print(baseConverter(32,16))

```

```python
from pythonds.basic import Stack

def infixToPostfix(infixToken):
    prec = {}
    prec["*"] = 3 
    prec["/"] = 3
    prec["-"] = 2
    prec["+"] = 2
    prec["("] = 1
    opStack = Stack()
    postFix = []
    tokens = infixToken.split()
    for token in tokens:
        if token in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" or token in "1234567890":
            postFix.append(token)
        elif token == "(":
            opStack.push(token)
        elif token == ")":
            topToken = opStack.pop()
            while topToken != "(":
                postFix.append(topToken)
                topToken = opStack.pop()
        else: 
            while (not opStack.isEmpty()) and (prec[opStack.peek()] >= prec[token]):
                postFix.append(opStack.pop())
            opStack.push(token)
    while not opStack.isEmpty():
        postFix.append(opStack.pop())
    return " ".join(postFix)
            
print(infixToPostfix("A * B + C * D"))
print(infixToPostfix("( A + B ) * C - ( D - E ) * ( F + G )"))
print(infixToPostfix("( A + B ) * ( C + D )"))
#'A B + C D + *'
print(infixToPostfix("( A + B ) * C"))
#'A B + C *'
print(infixToPostfix("A + B * C"))
#'A B C * +'
print(infixToPostfix("( A + B ) * ( C + D ) * ( E + F )"))
print(infixToPostfix(" A + ( ( B + C ) * ( D + E ) )"))
print(infixToPostfix("A * B * C * D + E + F"))
```

```python
from pythonds.basic import Stack

def postFixEval(postFixToken):
    tokens = postFixToken.split()
    operandStack = Stack()
    for token in tokens:
        if not token in "*-/+":
            operandStack.push(token)
        else:
            if not operandStack.isEmpty():
                op2 = int(operandStack.pop())
                op1 = int(operandStack.pop())
                result = doMath(op1,op2,token)
                operandStack.push(result)
    ret_result = operandStack.pop()
    if not operandStack.isEmpty():
        print("Did not work")
    return ret_result
        
def doMath(op1,op2,op3):
    if op3 == "*":
        result = op1 * op2
    elif op3 == "/":
        result = op1 / op2
    elif op3 == "+":
       result = op1 + op2
    else:
        result = op1 - op2
    return result
print(postFixEval('7 8 + 3 2 + /'))
print(postFixEval('10 3 5 * 16 4 - / +'))
print(postFixEval('2 3 * 4 +'))
print(postFixEval('1 2 + 3 + 4 + 5 +'))
print(postFixEval('1 2 3 4 5 * + * +'))
```

```python
"""
Simulation of the Hot Potato game

Description:
To begin, let’s consider the children’s game Hot Potato. 
In this game (see Figure 2) children line up in a circle and pass an item from neighbor to neighbor as fast as they can. 
At a certain point in the game, the action is stopped and the child who has the item (the potato) is removed from the circle. 
Play continues until only one child is left.
ex: 1, 2, 3, 4 uses
passes the hot potato to: ->
1 -> 2 then 1 pushed to the rear of the queue 
2 -> 3 then 2 pushed to the rear of the queue
3 -> 4 then 3 pushed to the rear of the queue
4 -> 1 then 4 pushed to the rear of the queue
pseudocode:
* list_of_players
* num_of_iterations_before_a_pop_action
* function hot_potato to simulate logic(list_of_players, num_of_iterations_before_a_pop_action)
    * initiate the queue
    * add the list of players to the queue
    * loop until one player is remaining
        * loop for the num_of_iterations_before_a_pop_action
    * return the remaining play
"""
from pythonds.basic import Queue

def hot_potato_game(list_of_players, num_of_iter):
    hot_potato_circle = Queue()
    for player in list_of_players:
        hot_potato_circle.enqueue(player)
    while hot_potato_circle.size() > 1:
        for i in range(num_of_iter):
            hot_potato_circle.enqueue(hot_potato_circle.dequeue())
        hot_potato_circle.dequeue()
    return hot_potato_circle.dequeue()
hot_potato_game(["Bill", "John", "Judith", "Peter", "Carlos", "Pirlo", "Arthur", "Sean", "Ruth", "David"], 12)
```

```python
"""
Consider the following situation in a computer science laboratory. On any average day about 10 students are working in the lab at any given hour.
These students typically print up to twice during that time, and the length of these tasks ranges from 1 to 20 pages. The printer in the lab is older,
capable of processing 10 pages per minute of draft quality. The printer could be switched to give better quality, 
but then it would produce only five pages per minute. The slower printing speed could make students wait too long. What page rate should be used?

If there are 10 students in the lab and each prints twice, 
then there are 20 print tasks per hour on average. What is the chance that at any given second, 
a print task is going to be created? The way to answer this is to consider the ratio of tasks to time. 
Twenty tasks per hour means that on average there will be one task every 180 seconds
Here is the main simulation.

Create a queue of print tasks. Each task will be given a timestamp upon its arrival. The queue is empty to start.

For each second (currentSecond):
Does a new print task get created? If so, add it to the queue with the currentSecond as the timestamp.
If the printer is not busy and if a task is waiting,
Remove the next task from the print queue and assign it to the printer.
Subtract the timestamp from the currentSecond to compute the waiting time for that task.
Append the waiting time for that task to a list for later processing.
Based on the number of pages in the print task, figure out how much time will be required.
The printer now does one second of printing if necessary. It also subtracts one second from the time required for that task.
If the task has been completed, in other words the time required has reached zero, the printer is no longer busy.
After the simulation is complete, compute the average waiting time from the list of waiting times generated
"""
from pythonds.basic import Queue
import random

class Printer:
    def __init__(self, ppm):
        self.page_rate = ppm
        self.task = None
        self.time_remaining = 0
    
    def tick(self):
        if self.task != None:
            self.time_remaining = self.time_remaining - 1
            if self.time_remaining <= 0:
                self.task = None
        
    def busy(self):
        if self.time_remaining > 0:
            return True
        return False
    def startNext(self, current_task):
        self.task = current_task
        self.time_remaining = ((current_task.get_pages() * 60)/self.page_rate)

class Task:
    def __init__(self, time):
        self.num_of_pages = random.randrange(1, 21)
        self.timestamp = time
    
    def get_pages(self):
        return self.num_of_pages
    
    def get_stamp(self):
        return self.timestamp
    
    def waiting_time(self, current_time):
        return current_time - self.timestamp

def newPrintTask():
    """Twenty tasks per hour means that on average there will be one task every 180 seconds"""
    if random.randrange(1, 181) != 180:
        return False
    return True

def simulationPrint(num_of_seconds_to_simulate, pages_per_min):
    labPrinter = Printer(pages_per_min)
    printQueue = Queue()
    waitingTimeOnQueue = []
    for second in range(num_of_seconds_to_simulate):
        if newPrintTask():
            PrintTask = Task(second)
            printQueue.enqueue(PrintTask)
        if (not labPrinter.busy()) and (not printQueue.isEmpty()):
            nextTask = printQueue.dequeue()
            waitingTimeOnQueue.append(nextTask.waiting_time(second))
            labPrinter.startNext(nextTask)
        labPrinter.tick()
    averageWait=sum(waitingTimeOnQueue)/len(waitingTimeOnQueue)
    print(f"Average Wait {averageWait:6.2f} secs {printQueue.size():3d} tasks remaining.")

for i in range(10):
    simulationPrint(3600, 5)
        
    
        
```

```python
"""Palindrome checker"""
from pythonds.basic import Deque

def palindromeChecker(word):
    wordChecked = Deque()
    for n in word:
        wordChecked.addFront(n)
    stillEqual = True
    while wordChecked.size() > 1 and stillEqual:
        charFront = wordChecked.removeFront()
        charRear = wordChecked.removeRear()
        if charFront != charRear:
            stillEqual = False
            
    return stillEqual

print(palindromeChecker("faljfdaslhfalhs"))
print(palindromeChecker("madam"))
print(palindromeChecker("root"))
print(palindromeChecker("toot"))
```
